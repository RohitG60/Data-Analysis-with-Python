### Write Python code to calculate

- Average mark across all students

- Average mark across male and female students

- Max and min mark

- Max mark for male and female students

- Average mark for class 'Five'

- Average mark for class 'Six' and male students only 

```python
# return average mark 
df["mark"].mean()
```

```python
# average for male students
df.loc[df["gender"]=="male"]["mark"].mean()
```

```python
# average for female students
df.query('gender=="female"')["mark"].mean()
```

```python
df["mark"].max()
```

```python
df["mark"].min()
```

```python
# max for male
df.loc[df["gender"]=="male"]["mark"].max()
```

```python
# max for female students
df.query('gender=="female"')["mark"].max()
```

```python
df.loc[df["class"]=="Five"]["mark"].mean()
```
